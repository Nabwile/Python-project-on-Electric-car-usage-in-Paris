# Checking for missing values

```python
# Creating a function to check
def missing_values(data):

  # identify the missing values from the Electric_cars dataset
  #
  miss = data.isnull().sum().sort_values(ascending = False)
  # the percentage of missing values
  percentage = (data.isnull().sum() / len(data)).sort_values(ascending = False)
  #storing  total missing values in percentage in dataframe
  #
  missing_data = pd.DataFrame({'Total' : miss, 'Percentage' : percentage})
  #if the percentage is 0, indicates no missing values hence removed
  #
  missing_data.drop(missing_data[missing_data['Percentage'] == 0].index, inplace = True)

  return missing_data

missing_values(Electric_cars)
```

```python
Electric_cars.columns
```

```python
# Creating a function to drop unnecessary columns
#
def drop_columns(data, cols):
  #drop columns not required
  data.drop(cols, axis = 1, inplace = True)

  return data

#columns to be dropped
columns = ['Cars', 'Displayed comment', 'Scheduled at', 'Charge Slots', 'Charging Status', 'Geo point', 'ID']
Electric_cars = drop_columns(Electric_cars, columns)
```

**Conclusion**


*   The column car was dropped since it contains similar values to those in bluecar counter. This is described in the documentation.
*   The columns Displayed comment and Scheduled had too many missing values
*   The column Geo point is used to provide location which is also provided in the column city
*  The columns Charge Slots and Charging Status are not required in the study
*   The column id is similar to address column therefore unnessecary

```python
# sort the data into ascending Datetime to enable there to be an order
#

from datetime import datetime

Electric_data = Electric_cars.sort_values(by='Datetime')
```

Sorted the Datetime column to clearly indicate dates that follow each other

```python
# Creating a column for bluecar usage
Electric_data['bluecar_diff'] = Electric_cars['bluecar counter'].diff()

# Creating a column for utilib usage
Electric_data['utilib_diff'] = Electric_cars['utilib counter'].diff()

# Creating a column for utilib 1.4 usage
Electric_data['utilib_two_diff'] = Electric_cars['utilib 1.4 counter'].diff()

# preview the table
Electric_data.head()
```

Adding a difference column for the three car counters (bluecar counter, utilib counter and utilib 1.4 counter)

+ve number means that a car was returned -ve number means that a car was picked 0 means that nothing happened, i.e no car was picked nor returned

# Analysis

```python
# Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018
# We selcet city == 'Paris' 

paris_data = Electric_data[Electric_data['city'] =='Paris'] 

paris_data.head()
```

```python
# What is the most popular hour for picking cars
# this lists the hour for picking cars
#

paris_data[(paris_data.bluecar_diff < 0)].groupby('hour')['hour'].count().sort_values(ascending = False).head(1);
```

```python
# What is the most popular hour for returning cars?
#this lists the hours for returning the cars 

paris_data[(paris_data.bluecar_diff > 0)].groupby('hour')['hour'].count().sort_values(ascending = False).head(1);
```

```python
# What station is the most popular overall?
# 

paris_data[(paris_data['kind'] == 'STATION') & (paris_data['status'] == 'ok')].groupby('public name')['public name'].count().sort_values(ascending= False).head(1);

```

```python
# What station is the most popular at the most popular picking hour?
#

paris_data[(paris_data['kind'] == 'STATION') & (paris_data['status'] == 'ok') & (paris_data['hour'] == 21)].groupby('public name')['public name'].count().sort_values(ascending= False).head(1);
```

```python
# What postal code is the most popular for picking up Blue cars overall? Does the most popular station belong to that postal code?

# First find the postal code of the most popular station overall
#
paris_data[paris_data['bluecar_diff'] < 0].groupby('postal code').count()[['bluecar_diff']].sort_values(by = 'bluecar_diff', ascending = False).head(1);

```

```python
# then cross check the postal code of the  most popular station
#

paris_data[paris_data['public name'] == 'Paris/Porte de Montrouge/8'][['public name', 'postal code']].head(1)
```

**Conclusion**

1.  The postal codes are not the same




```python
# What postal code is the most popular for picking up Blue cars at the most popular picking hour? 
# Does the most popular station belong to that postal code?
# 
# First find the postal code of the most popular station at most popular picking hour

paris_data[(paris_data.bluecar_diff < 0) & (paris_data.hour == 21)].groupby('postal code').count()[['bluecar_diff']].sort_values(by = 'bluecar_diff', ascending = False).head(1)

#paris_data[(paris_data['kind'] == 'STATION') & (paris_data['status'] == 'ok') & (paris_data['hour'] == 21)].groupby('postal code')['postal code'].count().sort_values(ascending= False).head(1)
```

```python
# Then cross check the name of the most popular station

paris_data[paris_data['public name'] == 'Paris/Porte de Montrouge/8'][['public name', 'postal code']].head(1)
```

**Conclusion**
*   The postal codes do not match
